```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import pandas as pd
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temps = session.query(Measurement.tobs).filter(func.strftime("%m", Measurement.date) == "05").all()
june_temps
```

```python
#  3. Convert the June temperatures to a list.
june_temps = [i[0] for i in june_temps]
june_temps
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_temps, columns=['June tobs'])
june_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_temps = session.query(Measurement.tobs).filter(func.strftime("%m", Measurement.date) == "12")
```

```python
# 7. Convert the December temperatures to a list.
december_temps = [i[0] for i in december_temps]
december_temps
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_df = pd.DataFrame(december_temps, columns=['December tobs'])
december_df.head()
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
december_df.describe()
```

```python
precip = session.query(Measurement.date, func.avg(Measurement.prcp)).\
    group_by(Measurement.date).order_by(Measurement.date).all()
precip = [i[1] for i in precip]
precip_df = pd.DataFrame(precip, columns=['Total precipitation'])
precip_df.describe()
```

```python
### Precipitation ###

# Since variable readings for each date, must get avg per day first then create df
june_precip = session.query(Measurement.date, func.avg(Measurement.prcp)).\
    filter(func.strftime("%m", Measurement.date) == "05").\
    group_by(Measurement.date).all()
june_precip = [i[1] for i in june_precip]
june_precip_df = pd.DataFrame(june_precip, columns=['June precipitation'])
june_precip_df.describe()
```

```python
dec_precip = session.query(Measurement.date, func.avg(Measurement.prcp)).\
    filter(func.strftime("%m", Measurement.date) == "12").\
    group_by(Measurement.date).all()
dec_precip = [i[1] for i in dec_precip]
dec_precip_df = pd.DataFrame(dec_precip, columns=['December precipitation'])
dec_precip_df.describe()
```
